```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows', None)
```

---

```python
file_path = "../datasets/wadi/WADI.A1_9 Oct 2017/WADI_14days.csv"
df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
print(df.dtypes)
print(df.shape)
```

```python
df.drop(["Row", "Date", "Time"], axis=1, inplace=True)
df.head()
```

```python
print(df.shape)
df.isnull().sum()
```

```python
df.dropna(thresh=df.shape[0]*0.6,how='all', axis=1, inplace=True)

print(df.shape)
df.isnull().sum()
```

```python
df.dropna(inplace=True)

print(df.shape)
df.isnull().sum()
```

---

```python
val_size = 0.3

train_df, val_test_df = train_test_split(df, test_size=val_size, random_state=42)
val_df, test_df = train_test_split(val_test_df, test_size=0.5, random_state=42)

train_data = train_df.to_numpy()
val_data = val_df.to_numpy()
test_data = test_df.to_numpy()

np.save("../datasets/wadi/wadi_train.npy", train_data)
np.save("../datasets/wadi/wadi_val.npy", val_data)
np.save("../datasets/wadi/wadi_test.npy", test_data)
```
